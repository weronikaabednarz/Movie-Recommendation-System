```python
# Importuje moduÅ‚ ast, ktÃ³ry jest potrzebny do przeksztaÅ‚cania stringÃ³w zawierajÄ…cych literaÅ‚y struktur danych Pythona 
# na rzeczywiste struktury danych Pythona.
import ast
```

```python
def convert(obj):   # Przyjmuje jeden argument obj, ktÃ³ry jest stringiem reprezentujÄ…cym listÄ™ sÅ‚ownikÃ³w.
    L = []          # Tworzy pustÄ… listÄ™ L, do ktÃ³rej zostanÄ… dodane nazwy z elementÃ³w przeksztaÅ‚conej listy sÅ‚ownikÃ³w.
    for i in ast.literal_eval(obj):     # wywoÅ‚uje funkcjÄ™ literal_eval() z moduÅ‚u ast, ktÃ³ra przeksztaÅ‚ca string obj na rzeczywistÄ… listÄ™
        L.append(i['name'])         #  Dla kaÅ¼dego sÅ‚ownika i w przeksztaÅ‚conej liÅ›cie, dodaje wartoÅ›Ä‡ pola 'name' do listy L.
    return L        #  Zwraca listÄ™ L, ktÃ³ra zawiera nazwy z przeksztaÅ‚conej listy sÅ‚ownikÃ³w
```

```python
movies_df['genres'] = movies_df['genres'].apply(convert)
movies_df['keywords'] = movies_df['keywords'].apply(convert)
movies_df.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])     # Dodaje nazwÄ™ aktualnego sÅ‚ownika do listy, do osiÄ…gniÄ™cia wartoÅ›ci 3
            counter +=1
        else:
            break
        return L        # zawiera nazwy z przetworzonych sÅ‚ownikÃ³w, ograniczone do trzech pierwszych
```

```python
movies_df['cast'] = movies_df['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
        return L
```

```python
movies_df['crew'] = movies_df['crew'].apply(fetch_director)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
def recommend(movie):
    movies_found = new_df[new_df['title'] == movie]
    if movies_found.empty:
        print("No recommendation for the introduced title.")
        return
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]
    recommended_movies = []
    for i in movies_list:
        recommended_movies.append(new_df.iloc[i[0]].title)
    return recommended_movies
```
